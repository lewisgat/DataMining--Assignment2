```python
from scipy.stats import entropy
from sklearn.decomposition import IncrementalPCA


H_mean = []
R_mean = []
G_mean = []
B_mean = []

H_var = []
R_var = []
G_var = []
B_var = []

H_ent = []
R_ent = []
G_ent = []
B_ent = []



for i in range(len(X)):
    
    ihc_hed = rgb2hed(X[i])
    null = np.zeros_like(ihc_hed[:, :, 0])
    ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))  #obtaining H channel from RGB values
    
    H_unique, H_count = np.unique(ihc_h, return_counts=True)  #counting the unique values in each H, R, G and B and returning their unique values and frequencies
    R_unique, R_count = np.unique(X[i,:,:,0], return_counts=True) 
    G_unique, G_count = np.unique(X[i,:,:,1], return_counts=True) 
    B_unique, B_count = np.unique(X[i,:,:,2], return_counts=True) 
    
    H_prob = H_count/len(H_count)    #frequncies used to calculate empirical probability
    R_prob = R_count/len(R_count)
    G_prob = G_count/len(G_count)
    B_prob = B_count/len(B_count) 
    
    H_mean.append(np.mean(ihc_h))   #average value of each H, R, G and B
    R_mean.append(np.mean(X[i,:,:,0])) 
    G_mean.append(np.mean(X[i,:,:,1]))
    B_mean.append(np.mean(X[i,:,:,2]))

    H_var.append(np.var(ihc_h))     #variance of each H, R, G and B
    R_var.append(np.var(X[i,:,:,0]))
    G_var.append(np.var(X[i,:,:,1]))
    B_var.append(np.var(X[i,:,:,2]))
    
    H_ent.append(entropy(H_prob))   #entropy of each H, R, G and B
    R_ent.append(entropy(R_prob))
    G_ent.append(entropy(G_prob))
    B_ent.append(entropy(B_prob))
    
    
    
```

```python
plt.rcParams['figure.figsize'] = [24, 18]

fig, axes = plt.subplots(4, 3)
ax = axes.ravel()

ax[0].scatter(np.array(Y['lymphocyte']), H_mean, color = 'darkorange', edgecolor='k')
ax[0].set_title("Lymphocyte Count vs H Channel Average")

ax[1].scatter(np.array(Y['lymphocyte']), H_var, color = 'orange', edgecolor='k')
ax[1].set_title("Lymphocyte Count vs H Channel Variance")

ax[2].scatter(np.array(Y['lymphocyte']), H_ent, color = 'goldenrod', edgecolor='k')
ax[2].set_title("Lymphocyte Count vs H Channel Entropy")

ax[3].scatter(np.array(Y['lymphocyte']), R_mean, color = 'indianred', edgecolor='k')
ax[3].set_title("Lymphocyte Count vs Red Colour Average")

ax[4].scatter(np.array(Y['lymphocyte']), R_var, color = 'maroon', edgecolor='k')
ax[4].set_title("Lymphocyte Count vs Red Colour Variance")

ax[5].scatter(np.array(Y['lymphocyte']), R_ent, color = 'red', edgecolor='k')
ax[5].set_title("Lymphocyte Count vs Red Colour Entropy")

ax[6].scatter(np.array(Y['lymphocyte']), G_mean, color = 'forestgreen', edgecolor='k')
ax[6].set_title("Lymphocyte Count vs Green Colour Average")

ax[7].scatter(np.array(Y['lymphocyte']), G_var, color = 'seagreen', edgecolor='k')
ax[7].set_title("Lymphocyte Count vs Green Colour Variance")

ax[8].scatter(np.array(Y['lymphocyte']), G_ent, color = 'lime', edgecolor='k')
ax[8].set_title("Lymphocyte Count vs Green Colour Entropy")

ax[9].scatter(np.array(Y['lymphocyte']), B_mean, color = 'blue', edgecolor='k')
ax[9].set_title("Lymphocyte Count vs Blue Colour Average")

ax[10].scatter(np.array(Y['lymphocyte']), B_var, color = 'royalblue', edgecolor='k')
ax[10].set_title("Lymphocyte Count vs Blue Colour Variance")

ax[11].scatter(np.array(Y['lymphocyte']), B_ent, color = 'navy', edgecolor='k')
ax[11].set_title("Lymphocyte Count vs Blue Colour Entropy")
```
