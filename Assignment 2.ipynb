```python
from scipy.stats import entropy
from sklearn.decomposition import IncrementalPCA
from sklearn.preprocessing import StandardScaler
```

```python
H_mean = []
R_mean = []
G_mean = []
B_mean = []

H_var = []
R_var = []
G_var = []
B_var = []

H_ent = []
R_ent = []
G_ent = []
B_ent = []

for i in range(len(X)):
    
    ihc_hed = rgb2hed(X[i]/255.0)
    null = np.zeros_like(ihc_hed[:, :, 0])
    ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))  #obtaining H channel from RGB values
    
    H_unique, H_count = np.unique(ihc_h, return_counts=True)  #counting the unique values in each H, R, G and B and returning their unique values and frequencies
    R_unique, R_count = np.unique(X[i,:,:,0], return_counts=True) 
    G_unique, G_count = np.unique(X[i,:,:,1], return_counts=True) 
    B_unique, B_count = np.unique(X[i,:,:,2], return_counts=True) 
    
    H_prob = H_count/len(H_count)    #frequncies used to calculate empirical probability
    R_prob = R_count/len(R_count)
    G_prob = G_count/len(G_count)
    B_prob = B_count/len(B_count) 
    
    H_mean.append(np.mean(ihc_h))   #average value of each H, R, G and B
    R_mean.append(np.mean(X[i,:,:,0])) 
    G_mean.append(np.mean(X[i,:,:,1]))
    B_mean.append(np.mean(X[i,:,:,2]))

    H_var.append(np.var(ihc_h))     #variance of each H, R, G and B
    R_var.append(np.var(X[i,:,:,0]))
    G_var.append(np.var(X[i,:,:,1]))
    B_var.append(np.var(X[i,:,:,2]))
    
    H_ent.append(entropy(H_prob))   #entropy of each H, R, G and B
    R_ent.append(entropy(R_prob))
    G_ent.append(entropy(G_prob))
    B_ent.append(entropy(B_prob))
```

```python
def data_setup(X_train, X_val, X_test):
    
    pca = IncrementalPCA(n_components=500, batch_size=512)
    scaler = StandardScaler()
    
    X_train_flat = X_train.flatten().reshape(len(y_train_reg), 3*256*256) 
    X_val_flat = X_val.flatten().reshape(len(y_val_reg), 3*256*256) 
    X_test_flat = X_test.flatten().reshape(len(y_test_reg), 3*256*256) 
    
    
    X_train_scale = scaler.fit_transform(X_train_flat)
    X_val_scale = scaler.transform(X_val_flat)
    X_test_scale = scaler.transform(X_test_flat)
    
    X_train_pca = pca.fit_transform(X_train_scale)
    X_val_pca = pca.transform(X_val_scale)
    X_test_pca = pca.transform(X_test_scale)
    
    H_mean_train = []
    R_mean_train = []
    G_mean_train = []
    B_mean_train = []

    H_var_train = []
    R_var_train = []
    G_var_train = []
    B_var_train = []

    H_ent_train = []
    R_ent_train = []
    G_ent_train = []
    B_ent_train = []
    
    H_mean_val = []
    R_mean_val = []
    G_mean_val = []
    B_mean_val = []

    H_var_val = []
    R_var_val = []
    G_var_val = []
    B_var_val = []

    H_ent_val = []
    R_ent_val = []
    G_ent_val = []
    B_ent_val = []

    H_mean_test = []
    R_mean_test = []
    G_mean_test = []
    B_mean_test = []

    H_var_test = []
    R_var_test = []
    G_var_test = []
    B_var_test = []

    H_ent_test = []
    R_ent_test = []
    G_ent_test = []
    B_ent_test = []


    for i in range(len(X_train)):

        ihc_hed = rgb2hed(X_train[i]/255)
        null = np.zeros_like(ihc_hed[:, :, 0])
        ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))  #obtaining H channel from RGB values

        H_unique, H_count = np.unique(ihc_h, return_counts=True)  #counting the unique values in each H, R, G and B and returning their unique values and frequencies
        R_unique, R_count = np.unique(X_train[i,:,:,0], return_counts=True) 
        G_unique, G_count = np.unique(X_train[i,:,:,1], return_counts=True) 
        B_unique, B_count = np.unique(X_train[i,:,:,2], return_counts=True) 

        H_prob = H_count/len(H_count)    #frequncies used to calculate empirical probability
        R_prob = R_count/len(R_count)
        G_prob = G_count/len(G_count)
        B_prob = B_count/len(B_count) 

        H_mean_train.append(np.mean(ihc_h))   #average value of each H, R, G and B
        R_mean_train.append(np.mean(X_train[i,:,:,0])) 
        G_mean_train.append(np.mean(X_train[i,:,:,1]))
        B_mean_train.append(np.mean(X_train[i,:,:,2]))

        H_var_train.append(np.var(ihc_h))     #variance of each H, R, G and B
        R_var_train.append(np.var(X_train[i,:,:,0]))
        G_var_train.append(np.var(X_train[i,:,:,1]))
        B_var_train.append(np.var(X_train[i,:,:,2]))

        H_ent_train.append(entropy(H_prob))   #entropy of each H, R, G and B
        R_ent_train.append(entropy(R_prob))
        G_ent_train.append(entropy(G_prob))
        B_ent_train.append(entropy(B_prob))
        
        
    for i in range(len(X_val)):

        ihc_hed = rgb2hed(X_val[i]/255)
        null = np.zeros_like(ihc_hed[:, :, 0])
        ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))  #obtaining H channel from RGB values

        H_unique, H_count = np.unique(ihc_h, return_counts=True)  #counting the unique values in each H, R, G and B and returning their unique values and frequencies
        R_unique, R_count = np.unique(X_val[i,:,:,0], return_counts=True) 
        G_unique, G_count = np.unique(X_val[i,:,:,1], return_counts=True) 
        B_unique, B_count = np.unique(X_val[i,:,:,2], return_counts=True) 

        H_prob = H_count/len(H_count)    #frequncies used to calculate empirical probability
        R_prob = R_count/len(R_count)
        G_prob = G_count/len(G_count)
        B_prob = B_count/len(B_count) 

        H_mean_val.append(np.mean(ihc_h))   #average value of each H, R, G and B
        R_mean_val.append(np.mean(X_val[i,:,:,0])) 
        G_mean_val.append(np.mean(X_val[i,:,:,1]))
        B_mean_val.append(np.mean(X_val[i,:,:,2]))

        H_var_val.append(np.var(ihc_h))     #variance of each H, R, G and B
        R_var_val.append(np.var(X_val[i,:,:,0]))
        G_var_val.append(np.var(X_val[i,:,:,1]))
        B_var_val.append(np.var(X_val[i,:,:,2]))

        H_ent_val.append(entropy(H_prob))   #entropy of each H, R, G and B
        R_ent_val.append(entropy(R_prob))
        G_ent_val.append(entropy(G_prob))
        B_ent_val.append(entropy(B_prob))

    for i in range(len(X_test)):

        ihc_hed = rgb2hed(X_test[i]/255)
        null = np.zeros_like(ihc_hed[:, :, 0])
        ihc_h = hed2rgb(np.stack((ihc_hed[:, :, 0], null, null), axis=-1))  #obtaining H channel from RGB values

        H_unique, H_count = np.unique(ihc_h, return_counts=True)  #counting the unique values in each H, R, G and B and returning their unique values and frequencies
        R_unique, R_count = np.unique(X_test[i,:,:,0], return_counts=True) 
        G_unique, G_count = np.unique(X_test[i,:,:,1], return_counts=True) 
        B_unique, B_count = np.unique(X_test[i,:,:,2], return_counts=True) 

        H_prob = H_count/len(H_count)    #frequncies used to calculate empirical probability
        R_prob = R_count/len(R_count)
        G_prob = G_count/len(G_count)
        B_prob = B_count/len(B_count) 

        H_mean_test.append(np.mean(ihc_h))   #average value of each H, R, G and B
        R_mean_test.append(np.mean(X_test[i,:,:,0])) 
        G_mean_test.append(np.mean(X_test[i,:,:,1]))
        B_mean_test.append(np.mean(X_test[i,:,:,2]))

        H_var_test.append(np.var(ihc_h))     #variance of each H, R, G and B
        R_var_test.append(np.var(X_test[i,:,:,0]))
        G_var_test.append(np.var(X_test[i,:,:,1]))
        B_var_test.append(np.var(X_test[i,:,:,2]))

        H_ent_test.append(entropy(H_prob))   #entropy of each H, R, G and B
        R_ent_test.append(entropy(R_prob))
        G_ent_test.append(entropy(G_prob))
        B_ent_test.append(entropy(B_prob))
        
    df_train = pd.DataFrame(X_train_pca)
    df_val = pd.DataFrame(X_val_pca)
    df_test = pd.DataFrame(X_test_pca)
        
    df_train['H Mean'] = H_mean_train
    df_train['H Var'] = H_var_train
    df_train['H Ent'] = H_ent_train
    df_train['R Mean'] = R_mean_train
    df_train['R Var'] = R_var_train
    df_train['R Ent'] = R_ent_train
    df_train['G Mean'] = G_mean_train
    df_train['G Var'] = G_var_train
    df_train['G Ent'] = G_ent_train
    df_train['B Mean'] = B_mean_train
    df_train['B Var'] = B_var_train
    df_train['B Ent'] = B_ent_train
    
    df_val['H Mean'] = H_mean_val
    df_val['H Var'] = H_var_val
    df_val['H Ent'] = H_ent_val
    df_val['R Mean'] = R_mean_val
    df_val['R Var'] = R_var_val
    df_val['R Ent'] = R_ent_val
    df_val['G Mean'] = G_mean_val
    df_val['G Var'] = G_var_val
    df_val['G Ent'] = G_ent_val
    df_val['B Mean'] = B_mean_val
    df_val['B Var'] = B_var_val
    df_val['B Ent'] = B_ent_val
    
    df_test['H Mean'] = H_mean_test
    df_test['H Var'] = H_var_test
    df_test['H Ent'] = H_ent_test
    df_test['R Mean'] = R_mean_test
    df_test['R Var'] = R_var_test
    df_test['R Ent'] = R_ent_test
    df_test['G Mean'] = G_mean_test
    df_test['G Var'] = G_var_test
    df_test['G Ent'] = G_ent_test
    df_test['B Mean'] = B_mean_test
    df_test['B Var'] = B_var_test
    df_test['B Ent'] = B_ent_test
    
    return df_train, df_val, df_test
```

```python
import torch
from torch.autograd import Variable
from torch.nn import *
from torch.optim import Adam, SGD
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import TensorDataset, DataLoader
import torchvision.transforms.functional as fn

train_x_reg_tensor = torch.tensor((df_train.values.astype(np.float32)))
val_x_reg_tensor = torch.tensor((df_val.values.astype(np.float32)))
test_x_reg_tensor = torch.tensor((df_test.values.astype(np.float32)))

train_y_reg_tensor = torch.tensor((y_train_reg.astype(np.float32)))
val_y_reg_tensor = torch.tensor((y_val_reg.astype(np.float32)))

train_y_reg_tensor = torch.unsqueeze(train_y_reg_tensor, 1)
val_y_reg_tensor = torch.unsqueeze(val_y_reg_tensor, 1)
```

```python
torch.cuda.empty_cache()


train_set = TensorDataset(train_x_reg_tensor, train_y_reg_tensor)
trainloader = torch.utils.data.DataLoader(
    train_set, 
    batch_size = 16,
    num_workers = 1,
    pin_memory = True, 
    shuffle=True)

val_set = TensorDataset(val_x_reg_tensor, val_y_reg_tensor)
validloader = torch.utils.data.DataLoader(
    val_set, 
    batch_size = 16,
    num_workers = 1,
    pin_memory = True, 
    shuffle=True)
```

```python
# Neural Network Model (2 hidden layers)
class MLP(Module):
    def __init__(self):
        super(MLP, self).__init__()

#         self.fc4 = Linear(200, 1)
        self.linear_layers = Sequential(
            Linear(513, 1000),
            ELU(),
            Dropout(0.2),
            Linear(1000, 2000),
            ELU(),
            Dropout(0.2),
            Linear(2000, 500),
            ELU(),
            Dropout(0.2),
            Linear(500, 1)
        )

    def forward(self, x):
        out = self.linear_layers(x)
        return out
    
```

```python
def train(epochs, print_score = False):   
    train_losses = [] 
    val_losses = []
    for e in range(epochs):
        train_loss = 0.0
        model.train()
        for data, cells in trainloader:
#             Transfer Data to GPU if available
            if torch.cuda.is_available():
                data, cells = data.cuda(), cells.cuda()
            data, cells = Variable(data), Variable(cells)
            # Clear the gradients
            optimizer.zero_grad()
            # Forward Pass
            target = model(data)
            # Find the Loss
            loss = criterion(target, cells)
            del target, cells
            # Calculate gradients
            loss.backward()
            # Update Weights
            optimizer.step()
            # Calculate Loss
            train_loss += loss.item()
            del loss
            
        valid_loss = 0.0
        model.eval()      
        for data, cells in validloader:
#             Transfer Data to GPU if available
            if torch.cuda.is_available():
                data, cells = data.cuda(), cells.cuda()
            data, cells = Variable(data), Variable(cells)
            # Forward Pass
            target = model(data)
            # Find the Loss
            loss = criterion(target, cells)
            del target, cells
            # Calculate Loss
            valid_loss += loss.item()
            del loss

        train_losses.append(np.sqrt(train_loss / len(trainloader)))
        val_losses.append(np.sqrt(valid_loss / len(validloader)))
       
        if  print_score == True:
            print('Epoch: ', e+1, '\t\t', ' Training RMSE:', round(np.sqrt(train_loss / len(trainloader)), 3), 
                  '\t\t', ' Val RMSE:', round(np.sqrt(valid_loss / len(validloader)), 3))
        
    
    return train_losses, val_losses    
```

```python
model = MLP()

torch.cuda.empty_cache()
# defining the optimizer
optimizer = Adam(model.parameters(), lr=0.0001) #rmse 20  at this lr
# defining the loss function
criterion = MSELoss()
if torch.cuda.is_available():
    criterion = criterion.cuda()
    model = model.cuda()
# defining the number of epochs
n_epochs = 25

# training the model
train_losses_MLP, val_losses_MLP = train(n_epochs, print_score = True)
```

```python
def train(epochs, print_score = False):   
    train_losses = [] 
    val_losses = []
    for e in range(epochs):
        train_loss = 0.0
        model.train()
        for data, cells in trainloader:
#             Transfer Data to GPU if available
            if torch.cuda.is_available():
                data, cells = data.cuda(), cells.cuda()
            data, cells = Variable(data), Variable(cells)
            # Clear the gradients
            optimizer.zero_grad()
            # Forward Pass
            target = model(data)
            # Find the Loss
            loss = criterion(target, cells)
            del target, cells
            # Calculate gradients
            loss.backward()
            # Update Weights
            optimizer.step()
            # Calculate Loss
            train_loss += loss.item()
            del loss
            
        valid_loss = 0.0
        model.eval()      
        for data, cells in validloader:
#             Transfer Data to GPU if available
            if torch.cuda.is_available():
                data, cells = data.cuda(), cells.cuda()
            data, cells = Variable(data), Variable(cells)
            # Forward Pass
            target = model(data)
            # Find the Loss
            loss = criterion(target, cells)
            del target, cells
            # Calculate Loss
            valid_loss += loss.item()
            del loss

        train_losses.append(np.sqrt(train_loss / len(trainloader)))
        val_losses.append(np.sqrt(valid_loss / len(validloader)))
        
#         if e == 0:
#             best_loss = np.sqrt(valid_loss / len(validloader))
#             torch.save(model.state_dict(), 'model_best_loss')
            
#         elif e > 0:
#             if np.sqrt(valid_loss / len(validloader)) < best_loss:
                
#                 best_loss = np.sqrt(valid_loss / len(validloader))
#                 torch.save(model.state_dict(), 'model_best_loss')

                
        if  print_score == True:
            print('Epoch: ', e+1, '\t\t', ' Training RMSE:', round(np.sqrt(train_loss / len(trainloader)), 3), 
                  '\t\t', ' Val RMSE:', round(np.sqrt(valid_loss / len(validloader)), 3))
        
    
    return train_losses, val_losses     
```

First, a simple CNN.

```python
class Net(Module):   
    def __init__(self):
        super(Net, self).__init__()

        self.cnn_layers = Sequential(
            # Defining a 2D convolution layer
            Conv2d(3, 12, kernel_size=5, stride=2),
            BatchNorm2d(12),
            LeakyReLU(),
            MaxPool2d(kernel_size=3, stride=1),
            Dropout(0.2),
            
            # Defining another 2D convolution layer
            Conv2d(12, 24, kernel_size=5, stride=2),
            BatchNorm2d(24),
            LeakyReLU(),
            MaxPool2d(kernel_size=2),
            Dropout(0.2),
            
            # Defining a third 2D convolution layer
            Conv2d(24, 48, kernel_size=3, stride=1),
            BatchNorm2d(48),
            LeakyReLU(),
            MaxPool2d(kernel_size=2, stride=1),
            Dropout(0.2),
            
            Conv2d(48, 96, kernel_size=3, stride=1),
            BatchNorm2d(96),
            LeakyReLU(),
            MaxPool2d(kernel_size=2, stride=1),
            Dropout(0.2),
            
            Flatten()
        )

        self.linear_layers = Sequential(
            Linear(6144  ,  5000),
            LeakyReLU(),
            Linear(5000 ,  1)
        )

    # Defining the forward pass    
    def forward(self, x):
        x = self.cnn_layers(x)
        x = x.view(x.size(0), -1)
        x = self.linear_layers(x)
        return x
```

There is an improvement in performance from the simple model but the model is slightly overfitting and the validation loss fluctuates through the epochs. In order to prevent the overfitting I will run the training using early stoppng, which will stop the training and return the model if the validation RMSE increases after it has run for a while. This early stopping monitoring occurs past a certain epoch, defined as 'patience' in order to allow for some initial fluctuation

```python
def train_early_stopping(epochs, print_score = False, patience = 5):   
    train_losses = [] 
    val_losses = []
    
    for e in range(epochs):
        train_loss = 0.0
        model.train()
        for data, cells in trainloader:
#             Transfer Data to GPU if available
            if torch.cuda.is_available():
                data, cells = data.cuda(), cells.cuda()
            data, cells = Variable(data), Variable(cells)
            # Clear the gradients
            optimizer.zero_grad()
            # Forward Pass
            target = model(data)
            # Find the Loss
            loss = criterion(target, cells)
            del target, cells
            # Calculate gradients
            loss.backward()
            # Update Weights
            optimizer.step()
            # Calculate Loss
            train_loss += loss.item()
            del loss
            
        valid_loss = 0.0
        model.eval()      
        for data, cells in validloader:
#             Transfer Data to GPU if available
            if torch.cuda.is_available():
                data, cells = data.cuda(), cells.cuda()
            data, cells = Variable(data), Variable(cells)
            # Forward Pass
            target = model(data)
            # Find the Loss
            loss = criterion(target, cells)
            del target, cells
            # Calculate Loss
            valid_loss += loss.item()
            del loss

        train_losses.append(np.sqrt(train_loss / len(trainloader)))
        val_losses.append(np.sqrt(valid_loss / len(validloader)))
        
        if e == 10:
            best_loss = np.sqrt(valid_loss / len(validloader))
            torch.save(model.state_dict(), 'best_model_params.pt')
            bad_streak = 0
            epoch_achieved = e
        elif e > 10:
            if best_loss > np.sqrt(valid_loss / len(validloader)):
                best_loss = np.sqrt(valid_loss / len(validloader))
                torch.save(model.state_dict(), 'best_model_params.pt')
                bad_streak = 0
                epoch_achieved = e
            else:
                bad_streak +=1

        
                
        if  print_score == True:
            print('Epoch: ', e+1, '\t\t', ' Training RMSE:', round(np.sqrt(train_loss / len(trainloader)), 3), 
                  '\t\t', ' Val RMSE:', round(np.sqrt(valid_loss / len(validloader)), 3))
            
        if e >= 10:
            if bad_streak >= patience:
                print('Early Stopping Activated')
                model.load_state_dict(torch.load('best_model_params.pt'))
                break
        
    
    return train_losses, val_losses, epoch_achieved
```

```python
prediction_arrays = {}
test_arrays = {}
RMSE_dict = {}
R2_dict = {}
pears_dict = {}
spear_dict = {}

for fold in folds:

    X_fold_train = X[F!= fold]
    X_fold_test = X[F == fold]

    y_fold_train = Y[F != fold]
    y_fold_test = Y[F == fold]

    train_x_fold = torch.from_numpy(X_fold_train/255.0).type(torch.float32)
    test_x_fold = torch.from_numpy(X_fold_test/255.0).type(torch.float32)
    
    for cell in Cells:
        train_y_fold = torch.tensor((y_fold_train[cell].values.astype(np.float32)))
        test_y_fold = (y_fold_test[cell].values.astype(np.float32))

        train_y_fold = torch.unsqueeze(train_y_fold, 1)

        train_x_fold = torch.permute(train_x_fold, (0, 3, 1, 2))
        test_x_fold = torch.permute(test_x_fold, (0, 3, 1, 2))

        train_x_fold_scaled = []
        for i in range(len(train_x_fold)):
            T = train_x_fold[i]
            T = T.unsqueeze(0)
            T = torch.nn.functional.interpolate(T,size=(128,128))
            T = T.squeeze(0)
            train_x_fold_scaled.append(T)

        test_x_fold_scaled = []
        for i in range(len(test_x_fold)):
            T = test_x_fold[i]
            T = T.unsqueeze(0)
            T = torch.nn.functional.interpolate(T,size=(128,128))
            T = T.squeeze(0)
            test_x_fold_scaled.append(T)

        train_x_fold_scaled = torch.stack(train_x_fold_scaled)  
        test_x_fold_scaled = torch.stack(test_x_fold_scaled)

        train_fold_set = TensorDataset(train_x_fold_scaled, train_y_fold)
        trainloader_fold = torch.utils.data.DataLoader(
        train_fold_set, 
        batch_size = 1,
        num_workers = 1,
        pin_memory = True, 
        shuffle=True)

        testloader_fold = torch.utils.data.DataLoader(
        test_x_fold_scaled, 
        batch_size = 1,
        num_workers = 1,
        pin_memory = True, 
        shuffle=True)
        
        del model
        gc.collect()
        torch.cuda.empty_cache()

        model_ft = models.vgg19_bn(pretrained=True)  

        model = Sequential(model_ft,  Linear(1000,1))

        del model_ft
        gc.collect()
        torch.cuda.empty_cache()

        # defining the optimizer

        optimizer = Adam(model.parameters(), lr=1e-3) 

        # defining the loss function
        criterion = MSELoss()

        if torch.cuda.is_available():
            criterion = criterion.cuda()
            model = model.cuda()
        # defining the number of epochs
        n_epochs = 25

        # training the model

        train_fold(n_epochs)

        torch.save(model.state_dict(), 'best_model_params_{}_fold{}.pt'.format(cell, fold))
        
        prediction_list = []
        with torch.no_grad():
            model.eval()
            model = model.cuda()
            for data in testloader_fold:
                data = data.cuda()
                data = Variable(data)
                pred = model(data)
                prediction_list.append(torch.squeeze(pred, 1).cpu().numpy())

        y_test_pred_CNN_fold = np.squeeze(np.asarray(prediction_list))

        RMSE_CNN_fold = np.sqrt(mean_squared_error(y_test_pred_CNN_fold, test_y_fold))
        R2_CNN_fold = r2_score(y_test_pred_CNN_fold, test_y_fold)
        spearman_coef_CNN_fold = stats.spearmanr(y_test_pred_CNN_fold, test_y_fold)
        pearson_coef_CNN_fold = stats.pearsonr(y_test_pred_CNN_fold, test_y_fold)
        
        name = '{}_fold_{}'.format(cell, fold)
        prediction_arrays[name]= y_test_pred_CNN_fold
        test_arrays[name] = test_y_fold
        RMSE_dict[name]=RMSE_CNN_fold
        R2_dict[name]= R2_CNN_fold
        pears_dict[name]= pearson_coef_CNN_fold
        spear_dict[name]= spearman_coef_CNN_fold
```
